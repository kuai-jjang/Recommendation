```python
haha=pd.read_csv('./lecture_vector_2018_2.csv')
```

```python
dict(zip(haha[haha.columns[0]],haha[haha.columns[2:]].values))
```

```python
data=haha[haha.columns[1:]]
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

tsne_data = TSNE(learning_rate=300, init='pca').fit_transform(data)

# data=pd.DataFrame(haha).T
# pca = PCA(n_components=2)
# y = pca.fit_transform(data)

# data['pca_x']=y[:,0]
# data['pca_y']=y[:,1]
```

```python
data = (
  pd.DataFrame(tsne_data, columns=['V1', 'V2'])
    .assign(classname = haha[haha.columns[0]])
)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib import font_manager,rc

font_path = "C:/Windows/Fonts/HMFMPYUN.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)

plt.figure(figsize=(10,10))
for i,an in enumerate(data.index.values):

    x = data['V1'].iloc[i]
    y = data['V2'].iloc[i]
    plt.scatter(x, y, marker='x', color='blue')
    if (i+5)%100==0:
        plt.text(x+0.01, y+0.01, data['classname'].iloc[i], fontsize=10,color='black')
        
plt.show()

```
