```python
import pickle
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from matplotlib import font_manager,rc

def nplot(freq,n=10):
    freq_sort=sorted(freq.items(),key=lambda x:x[1],reverse=True)
    x, y = zip(*freq_sort) # unpack a list of pairs into two tuples
    z,_=zip(*x)
    sns.barplot(list(z[:n]),list(y[:n]))
    plt.title('전처리 후')
    plt.figure(figsize=(10,10))
    
    #return z,y




font_path = "C:/Windows/Fonts/HMFMPYUN.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)



with open('./vocab_without_josa_gut_su.pickle','rb') as f:
    w2i=pickle.load(f)
with open('./vocab_without_josa_gut_su_freq.pickle','rb') as f:
    freq=pickle.load(f)

    
nplot(freq)
```
